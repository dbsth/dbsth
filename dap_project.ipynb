```python
import pandas as pd
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/dbsth/dbsth/main/data/theater')
df
```

```python
del df['위도']
del df['경도']
del df['인터미션']
del df['건물 번호']
del df['최종작성일']
del df['법정읍면동명칭']
del df['출연진']
del df['리 명칭']
del df['제목']
del df['공연 시작일']
del df['공연 마지막일']
del df['도로명 이름']
del df['도로명주소']
del df['번지']
del df['우편번호']
del df['지번주소']

df
```

2. 공연 별 시간 구분
- 문화 종류 별 시간 비교
3. 시도 별 공연 빈도 비교
- 시도 별 문화 종류 빈도 비교

```python
df['시간'] = df['시간'].str.replace('분','')
df['시간'] = df['시간'].astype("float")
df
```

```python
df3 = df.groupby(df['시도 명칭'])
df3_1 = df3.get_group('강원도')
df3_1.rename(columns={'연령':'강원도'}, inplace=True)
df3_2 = df3.get_group('경기도')
df3_2.rename(columns={'연령':'경기도'}, inplace=True)
df3_3 = df3.get_group('경상남도')
df3_3.rename(columns={'연령':'경상남도'}, inplace=True)
df3_4 = df3.get_group('경상북도')
df3_4.rename(columns={'연령':'경상북도'}, inplace=True)
df3_5 = df3.get_group('광주광역시')
df3_5.rename(columns={'연령':'광주광역시'}, inplace=True)
df3_6 = df3.get_group('대구광역시')
df3_6.rename(columns={'연령':'대구광역시'}, inplace=True)
df3_7 = df3.get_group('대전광역시')
df3_7.rename(columns={'연령':'대전광역시'}, inplace=True)
df3_8 = df3.get_group('부산광역시')
df3_8.rename(columns={'연령':'부산광역시'}, inplace=True)
df3_9 = df3.get_group('서울특별시')
df3_9.rename(columns={'연령':'서울특별시'}, inplace=True)
df3_10 = df3.get_group('세종특별자치시')
df3_10.rename(columns={'연령':'세종특별자치시'}, inplace=True)
df3_11 = df3.get_group('울산광역시')
df3_11.rename(columns={'연령':'울산광역시'}, inplace=True)
df3_12 = df3.get_group('인천광역시')
df3_12.rename(columns={'연령':'인천광역시'}, inplace=True)
df3_13 = df3.get_group('전라남도')
df3_13.rename(columns={'연령':'전라남도'}, inplace=True)
df3_14 = df3.get_group('전라북도')
df3_14.rename(columns={'연령':'전라북도'}, inplace=True)
df3_15 = df3.get_group('제주특별자치도')
df3_15.rename(columns={'연령':'제주특별자치도'}, inplace=True)
df3_16 = df3.get_group('충청남도')
df3_16.rename(columns={'연령':'충청남도'}, inplace=True)
df3_17 = df3.get_group('충청북도')
df3_17.rename(columns={'연령':'충청북도'}, inplace=True)
```

```python
df3_1 = df3_1['강원도'].groupby(df['문화종류']).count()
df3_2 = df3_2['경기도'].groupby(df['문화종류']).count()
df3_3 = df3_3['경상남도'].groupby(df['문화종류']).count()
df3_4 = df3_4['경상북도'].groupby(df['문화종류']).count()
df3_5 = df3_5['광주광역시'].groupby(df['문화종류']).count()
df3_6 = df3_6['대구광역시'].groupby(df['문화종류']).count()
df3_7 = df3_7['대전광역시'].groupby(df['문화종류']).count()
df3_8 = df3_8['부산광역시'].groupby(df['문화종류']).count()
df3_9 = df3_9['서울특별시'].groupby(df['문화종류']).count()
df3_10 = df3_10['세종특별자치시'].groupby(df['문화종류']).count()
df3_11 = df3_11['울산광역시'].groupby(df['문화종류']).count()
df3_12 = df3_12['인천광역시'].groupby(df['문화종류']).count()
df3_13 = df3_13['전라남도'].groupby(df['문화종류']).count()
df3_14 = df3_14['전라북도'].groupby(df['문화종류']).count()
df3_15 = df3_15['제주특별자치도'].groupby(df['문화종류']).count()
df3_16 = df3_16['충청남도'].groupby(df['문화종류']).count()
df3_17 = df3_17['충청북도'].groupby(df['문화종류']).count()

df4_1 = pd.concat([df3_1, df3_2], axis=1)
df4_2 = pd.concat([df3_3, df3_4], axis=1)
df4_3 = pd.concat([df3_5, df3_6], axis=1)
df4_4 = pd.concat([df3_7, df3_8], axis=1)
df4_5 = pd.concat([df3_9, df3_10], axis=1)
df4_6 = pd.concat([df3_11, df3_12], axis=1)
df4_7 = pd.concat([df3_13, df3_14], axis=1)
df4_8 = pd.concat([df3_15, df3_16], axis=1)

df4_1 = pd.concat([df4_1, df4_2], axis=1)
df4_2 = pd.concat([df4_3, df4_4], axis=1)
df4_3 = pd.concat([df4_5, df4_6], axis=1)
df4_4 = pd.concat([df4_7, df4_8], axis=1)

df4_1 = pd.concat([df4_1, df4_2], axis=1)
df4_2 = pd.concat([df4_3, df4_4], axis=1)

df4_1 = pd.concat([df4_1, df4_2], axis=1)

df4 = pd.concat([df4_1, df3_17], axis=1)
df4 = df4.fillna('0')
df4
```
